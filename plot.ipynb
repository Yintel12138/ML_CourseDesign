```python
import numpy as np
import pandas as pd

df = pd.read_csv("result.csv",header=None)
np.array(df[0:6][1])
```

```python
import matplotlib.pyplot as plt

x = range(6)
x_label = ['Precision', 'Recall', 'F1-Score']
yt = ["","","SVM","SVM","",""]
color = ["r","g","b"]

for i in range(0,3):
  plt.figure(dpi=150)
  plt.grid(False)
  plt.title('Iris')  # 
  plt.xticks(x,yt)
  plt.rcParams['font.sans-serif'] = ["SimSum",'SimHei']  # 
  plt.xlabel('')  # x
  plt.ylabel('')  # y
  for index,item in enumerate(df[0:6][i]):
    plt.text(x[index], item, round(item,6), ha='center', va='bottom', fontsize=8)
  plt.plot(x, df[0:6][i], marker='o', markersize=3,label=x_label[i],color=color[i])
  plt.legend()  # 
  plt.show()  # 
```
