```python
import numpy as np
import pandas as pd

df = pd.read_csv("result.csv",header=None)
np.array(df[0:6][1])
```

```python
import matplotlib.pyplot as plt

x = range(6)
x_label = ['精确率Precision', '召回率Recall', 'F1-Score']
yt = ["朴素贝叶斯","逻辑回归","SVM（线性）","SVM（高斯）","决策树","随机森林"]
color = ["r","g","b"]

for i in range(0,3):
  plt.figure(dpi=150)
  plt.grid(False)
  plt.title('Iris数据集')  # 折线图标题
  plt.xticks(x,yt)
  plt.rcParams['font.sans-serif'] = ["SimSum",'SimHei']  # 显示汉字
  plt.xlabel('算法种类')  # x轴标题
  plt.ylabel('指标')  # y轴标题
  for index,item in enumerate(df[0:6][i]):
    plt.text(x[index], item, round(item,6), ha='center', va='bottom', fontsize=8)
  plt.plot(x, df[0:6][i], marker='o', markersize=3,label=x_label[i],color=color[i])
  plt.legend()  # 设置折线名称
  plt.show()  # 显示折线图
```
